```python
def remove_currency(x):
    return x.split(' ')[1]

def clean_columns(column):
    _list = []
    for x in column:
        a = x.lower().replace('(', ' ').replace(")", '').split(' ')
        a = '_'.join(a)
        _list.append(a)

    return _list


def get_date(path):
    print(path)
    return datetime.strptime(path.split('/')[-1].split('.')[0],"%m-%d-%Y")


def transform(dataset, date):
    try:
        if dataset['commodities'].iloc[0] == "No data available in table":
            dataset['commodities'] = np.nan
    
        try:
            dataset['average'] = dataset['average'].apply(lambda a: remove_currency(a))
        except AttributeError:
            pass
    
        
        dataset = dataset[['commodities',  'average']]
        
        
    
        dataset = dataset.T
        dataset.columns = clean_columns(list(dataset.iloc[0]))
        dataset = dataset[1:]
        dataset.reset_index(drop = True)

        date = get_date(date)
        dataset['date'] = date
        dataset.set_index('date', inplace = True, drop = True)
        return dataset
    except:
        date = get_date(date)
        throwaway_dataset = {'no_data_available': True}
        dataset = pd.DataFrame(throwaway_dataset)
        dataset['date'] = date
        dataset.set_index('date', inplace = True, drop = True)
        return dataset
    
```

```python
final_dataset = None
for i in range(5):
    if i == 0:
        dataset = pd.read_csv(random_samples[i])
        dataset = transform(dataset, random_samples[i])
        final_dataset = dataset
    else:
        dataset = pd.read_csv(random_samples[i])
        dataset = transform(dataset, random_samples[i])
        final_dataset = pd.concat([final_dataset, dataset], ignore_index = False, sort = False)
```
